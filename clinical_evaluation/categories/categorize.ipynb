```python

# keywords = {
#     'trung_that': ['trung thất', 'khí quản đoạn cao', 'phế quản', 'khí quản', 'cửa sổ', 'carina', 'động mạch chủ', 'mạch máu', 'rốn phổi', 'hoành', 'thực quản' ], 
#     'phoi': ['thùy', 'nhu mô', 'phổi', 'phổi' 'phổi trái', 'phổi phải', 'màng phổi' ],
#     'bung': ['bụng', 'tâm vị'],
#     'cổ': ['hạch cổ', 'cằm', 'hàm', 'thượng đòn', 'hạ đòn'],
#     'nách': ['nách'], 
# }
keywords = {
    'trung_that': ['trung thất', 'khí quản đoạn cao', 'phế quản', 'khí quản', 'cửa sổ', 'carina', 'động mạch chủ', 'mạch máu', 'rốn phổi', 'hoành', 'thực quản', 'đoạn thấp' ], 
    'phoi': ['thùy', 'nhu mô', 'phổi', 'phổi' 'phổi trái', 'phổi phải', 'màng phổi' ],
    'bung': ['bụng', 'tâm vị', 'dạ dày', 'dạ dài', 'trực tràng'],
    'cổ': ['hạch cổ', 'cằm', 'hàm', 'thượng đòn', 'hạ đòn', 'cơ ức đòn chũm', 'nhóm iib trái', 'nhóm vi bên phải', 'nhóm'],
    'nách': ['nách'], 
}


def check(keywords, text, key): 
    for kw in keywords[key]:
        if kw in text :
            return True 
    return False

def soft_rules_level_1(set_of_texts): 
    cattegories = {}
    cattegories['khac'] = []
    for key in keywords.keys(): 
        cattegories[key] = []
    for text in set_of_texts:
        khac = True
        for key in keywords.keys(): 
            if check(keywords, text, key): 
                cattegories[key].append(text)
                khac = False
                break 
        if khac: 
            cattegories['khac'].append(text)
    return cattegories
```

```python
categories = soft_rules_level_1(vitri_gt)
categories
```

```python
with io.open('gt_vitri_new.json', 'w', encoding='utf-8') as f:
    json.dump(categories, f, ensure_ascii=False, indent=4)
```
